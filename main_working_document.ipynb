```python
# By writing this as a function, we don't have to repeat it later. It's less code, it makes
# the rest easier to read, and if we improve this, we have to do it only once!

letters_to_numbers = {
    'A': 0,
    'B': 1,
    'C': 2,
    'D': 3,
    'E': 4,
    'F': 5,
    'G': 6,
    'H': 7,
    'I': 8,
    'J': 9
}

def ask_user_for_board_position():
    column = input("column (A to J):")
    while column.upper() not in "ABCDEFGHIJ":
        print("That column is wrong! It should be A, B, C, D, E, F, G, H, I or J")
        column = input("column (A to J):").upper()

    row = input("row (1 to 10):")
    while row not in "12345678910":
        print("That row is wrong! it should be 1, 2, 3, 4, 5, 6, 7, 8, 9 or 10")
        row = input("row (1 to 10):")

    # The code calling this function will receive the values listed in the return statement below
    # and it can assign it to variables
    return int(row) - 1, letters_to_numbers[column]
```

```python
def check_space(map):
    for row in range(1,11):
        for column in range(1,11):
            return map[row][column] == " "
```

```python
# define the shooting behavior
def shoot(shoot_map,shoot_position):
    row, column = shoot_position
    # "●" marker represents the shoot 
    shoot_map[row][column] = "●"
```

```python
# check if the guess point hit the ship or not
# Input argument:
# 1. shoot_map: the map for shooting
# 2. shoot_position: your shoot position 
# 3. home_position_oppent: your opponent's home position
def check_hit_miss(shoot_map,shoot_position,home_map_opponent):
    row, column = shoot_position
    if home_map_opponent[row][column] == "0":
        home_map_opponent[row][column] = "X"
        print("Yes, you hit it!")
       
        # put "●" marker into the surrounding grids 
        shoot_map[row-1][column-1] = "●"
        shoot_map[row-1][column] = "●"
        shoot_map[row-1][column+1] = "●"

        shoot_map[row][column-1] = "●"
        shoot_map[row][column+1] = "●"
        
        shoot_map[row+1][column-1] = "●"
        shoot_map[row+1][column] = "●"
        shoot_map[row+1][column+1] = "●"

        return True
    else:
        print("No, you miss it!")
        return False
```

```python
# Check if all the ships get sunk

sunk_ship_dic = [{"name": "Player1", "sink_position": [],"sink_ship_num":None},   # For player 1: (list of positions, list of positions, sunk ship number)
                 {"name": "Player2", "sink_position": [],"sink_ship_num":None}   # For player 2: (list of positions, list of positions, sunk ship number)
                ]


def check_win(player,sunk_ship_dic,ship_num):
    #print("Now we are checking!")
    if player == "Player1":
        if sunk_ship_dic[0]["sink_ship_num"] == ship_num:
           # print("Player1 wins!")
            return True
    else:
        if sunk_ship_dic[1]["sink_ship_num"] == ship_num:
            #print("Player2 wins!")
            return True

    return False
```

```python
#Claclulate if entire ship is sunk and check how many ships are sunk

def ship_sunk(player,shoot_map, home_map_opponent, sunk_ship_dic):
    # Find the sunk ship based on the shoot map and the opponent's home map
    if player == "Player1":
        for i in range(len(shoot_map)):
            for j in range(len(shoot_map[i])):
                if shoot_map[i][j] == "●" and home_map_opponent[i][j] == "X":
                    sunk_ship_position = home_map_opponent[i][j]  # Assuming the ship type is marked in the home map
                    if sunk_ship_position not in sunk_ship_dic[0]["sink_position"]:
                        sunk_ship_dic[0]["sink_ship_num"] += 1
                        sunk_ship_dic[0]["sink_position"].append(sunk_ship_position)  # Add the sunk ship number if not already presen
                    return sunk_ship_dic[0]  # Return the sunk ship number for confirmation
    else:
         for i in range(len(shoot_map)):
            for j in range(len(shoot_map[i])):
                if shoot_map[i][j] == "●" and home_map_opponent[i][j] == "X":
                    sunk_ship_position = home_map_opponent[i][j]  # Assuming the ship type is marked in the home map
                    if sunk_ship_position not in sunk_ship_dic[1]["sink_position"]:
                        sunk_ship_dic[1]["sink_ship_num"] += 1
                        sunk_ship_dic[1]["sink_position"].append(sunk_ship_position)  # Add the sunk ship number if not already presen
                    return sunk_ship_dic[1]  # Return the sunk ship number for confirmation
    return None  # If no ship is sunk, return None
```

## Game Logic

```python
# Logic of all the functions run together:

sunk_ship_dic = [{"name": "player1", "sink_position": [],"sink_ship_num":None},   # For player 1: (list of positions, list of positions, sunk ship number)
                 {"name": "player2", "sink_position": [],"sink_ship_num":None}   # For player 2: (list of positions, list of positions, sunk ship number)
                ]

ship_shape = "0"
ship_num = 1

while True:

    # Choose randomly who is the first
    turn = choose_turn()
    print(f"{turn} will start!")

    # There exist 4 maps in total:  
    #For player1: a home map, a shooting map;  
    home_map_1 = [[' ' for j in range(10)] for i in range(10)]
    shoot_map_1= [[' ' for j in range(10)] for i in range(10)]

    #For player2: a home map, a shooting map;   
    home_map_2 = [[' ' for j in range(10)] for i in range(10)]
    shoot_map_2= [[' ' for j in range(10)] for i in range(10)]

    # Ask player 1 to place the ship on the home map in the desired home position
    print("Place your ship")
    display_grid("home","Player1",home_map_1)
    home_position_1 = ask_user_for_board_position()
    place_ship(home_map_1, ship_shape, home_position_1)
    

    # Ask player 2 to place the ship on the home map in the desired home position
    print("Place your ship")
    display_grid("home","Player2",home_map_2)
    home_position_2 = ask_user_for_board_position()
    place_ship(home_map_2, ship_shape, home_position_2)
    

    # Make sure the player is ready
    play_game = input("Are you ready? Y/N: ")
    if play_game.lower().startswith("y"):
        game_on = True
    else:
        game_on = False

    # logic for player1's turn
    while game_on:
        print("Now the game is on!")
        if turn == "Player1":
            print("Now it's Player1's turn! Please choose your shot position!")
            shoot_position = ask_user_for_board_position()
            shoot(shoot_map_1, shoot_position)
            display_grid("shoot", "Player1", shoot_map_1)

            hit_miss = check_hit_miss(shoot_map_1, shoot_position, home_map_2)
            if hit_miss:
                print("It's a hit!")
                if check_win("Player1", sunk_ship_dic, ship_num):
                    print("Congratulations! Player1 wins the game!")
                    break
                else:
                    # Player 1 shoots the map again on the desired shoot position
                    shoot_position_1 = ask_user_for_board_position()
                    shoot(shoot_map_1,shoot_position_1)
                    display_grid("shoot", "Player1", shoot_map_1)
            else:
                
                print("It's a miss!")

                # Switch turn
            turn = "Player2"

        else:  # it's Player2's turn
            print("Now it's Player2's turn! Please choose your shot position!")
            shoot_position = ask_user_for_board_position()
            shoot(shoot_map_2, shoot_position)
            display_grid("shoot", "Player2", shoot_map_2)

            hit_miss = check_hit_miss(shoot_map_2, shoot_position, home_map_1)
            if hit_miss:
                print("It's a hit!")
                if check_win("Player2", sunk_ship_dic, ship_num):
                    print("Congratulations! Player2 wins the game!")
                else:
                    # Player 2 shoots the map again on the desired shoot position
                    shoot_position_2 = ask_user_for_board_position()
                    shoot(shoot_map_2,shoot_position_1)
                    display_grid("shoot", "Player2", shoot_map_2)
            else:
                print("It's a miss!")

            # Switch turn
            turn = "Player1"
```
